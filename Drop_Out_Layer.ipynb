##During training, dropout randomly "drops" (turns off) some neurons in a layer for each training batch.
### * This means those neurons do not participate in forward or backward propagation.

### * The dropped neurons are chosen randomly every time.

##Because different sets of neurons are active for each batch:

### * The model learns different combinations of features.

### * It prevents any single neuron from becoming too important.

### *This forces the network to become more robust and generalize better.

```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/sonar_dataset.csv",header=None)
```

```python
df.sample(10)
```

```python
df.columns
```

```python
df[60].value_counts()
```

```python
x=df.drop(60,axis=1)
```

```python
y=df[60]
```

```python
y=pd.get_dummies(y,drop_first=True)
```

```python
y.sample(5)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=25,random_state=1)
print(x_train.shape)
print(x_test.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
model=keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,)),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=50,batch_size=8)
```

```python
model.evaluate(x_test,y_test)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
y_pred=model.predict(x_test).reshape(-1)
y_pred=np.round(y_pred)
acc=accuracy_score(y_test,y_pred)
print(acc)

```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

#Adding DropOut layers

```python
import tensorflow as tf
from tensorflow import keras
model=keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,)),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
     keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'),
     keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=50,batch_size=8)
```

```python
model.evaluate(x_test,y_test)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
y_pred=model.predict(x_test).reshape(-1)
y_pred=np.round(y_pred)
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
